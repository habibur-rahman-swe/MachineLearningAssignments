```python
import pandas as pd
import pylab as pl
import numpy as np
import scipy.optimize as opt
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
%matplotlib inline
import matplotlib.pyplot as plt

from sklearn.metrics import classification_report
from sklearn import svm
from sklearn.metrics import f1_score
```

```python
def wragle(path):
  df = pd.read_csv(path)

  # drop index, Patient Id
  df.drop(columns=['index', 'Patient Id'], inplace=True)

  # changing the level low = 0, medium = 1, high = 1
  df['Level'] = df['Level'].map({'Low': 0, 'Medium': 1, 'High': 1})

  return df
```

```python
df =  wragle("/content/sample_data/cancer_patient_data_sets.csv")
df.head()
```

```python
X = df.drop(columns=['Level'])
y = df['Level']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=4, stratify=y)
print('Train set: ', X_train.shape, y_train.shape)
print('Test set: ', X_test.shape, y_test.shape)
```

```python
kernels = {
    'rbf': 'Radial Basis Function',
    'linear': 'Linear Function',
    'poly': 'Polynomial Function',
    'sigmoid': 'Sigmoid Function'
}

result = {
    'Model': [],
    'F1-score': []
}

def for_kernel_gc(kernel):
  max_f1 = 0
  max_g = 0
  max_c = 0

  for g in range(1, 11):
    for c in range(1, 11):
      f1 = tune_model(kernel, g, c)
      if f1 > max_f1:
        max_f1 = f1
        max_g = g
        max_c = c
      if max_f1 == 1:
        break
    if max_f1 == 1:
      break
  return [max_g, max_c]

def tune_model(kernel, gamma,  c):
  clf = svm.SVC(kernel=kernel, gamma=gamma, C=c)
  clf.fit(X_train, y_train)
  yhat = clf.predict(X_test)
  # print(kernels[kernel], ", Gamma:",  gamma, ", C:", c , ", Avg F1-score: %.4f" % f1_score(y_test, yhat, average='weighted'))
  return f1_score(y_test, yhat, average='weighted')
```

# Modeling svm
Radial basis function

1.   Radial basis function



```python
kernel = 'rbf'
g, c = for_kernel_gc(kernel)

clf = svm.SVC(kernel=kernel, gamma=g, C=c)
clf.fit(X_train, y_train)
yhat = clf.predict(X_test)
print("Avg F1-score: %.4f" % f1_score(y_test, yhat, average='weighted'))

result['F1-score'].append(f1_score(y_test, yhat, average='weighted'))
result['Model'].append(kernels[kernel])
```

2.   Linear function

```python
kernel = 'linear'
g, c = for_kernel_gc(kernel)
clf = svm.SVC(kernel=kernel, gamma=g, C=c)
clf.fit(X_train, y_train)
yhat = clf.predict(X_test)
print("Avg F1-score: %.4f" % f1_score(y_test, yhat, average='weighted'))

result['F1-score'].append(f1_score(y_test, yhat, average='weighted'))
result['Model'].append(kernels[kernel])
```

3. Polinomial

```python
kernel = 'poly'
g, c = for_kernel_gc(kernel)

clf = svm.SVC(kernel=kernel, gamma=g, C=c)
clf.fit(X_train, y_train)
yhat = clf.predict(X_test)
print("Avg F1-score: %.4f" % f1_score(y_test, yhat, average='weighted'))

result['F1-score'].append(f1_score(y_test, yhat, average='weighted'))
result['Model'].append(kernels[kernel])
```

4. Sigmoid

```python
kernel = 'sigmoid'
g, c = for_kernel_gc(kernel)

clf = svm.SVC(kernel=kernel, gamma=g, C=c)
clf.fit(X_train, y_train)
yhat = clf.predict(X_test)
print("Avg F1-score: %.4f" % f1_score(y_test, yhat, average='weighted'))

result['F1-score'].append(f1_score(y_test, yhat, average='weighted'))
result['Model'].append(kernels[kernel])
```

```python
result_df = pd.DataFrame(result)
result_df.set_index('Model', inplace=True)
result_df
```
